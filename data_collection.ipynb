```python
#Web Scraping-->Extracting data avaailable on Websites
import requests

URL="https://www.scrapethissite.com/pages/simple/"
res=requests.get(URL)
```

```python
print(res.status_code)
```

```python
if (res.status_code==200):
    # print(res.text)
    # print(res.content)
    print(res.headers) #MetaData like content type,server info etc


with open("scraped_data/data1.html","w", encoding="utf-8") as f:
    f.write(res.text)
```

```python
#BeautifulSoup-->It is a Python library to parse html and xml documents.Helps to extract data from webpages

!pip install beautifulsoup4
```

```python
from bs4 import BeautifulSoup

with open("scraped_data/data1.html","r",encoding="utf-8") as f:
    html_content=f.read()


soup=BeautifulSoup(html_content,"lxml") #lxm-->html parser
```

```python
soup.find("h1") #find the first matching element
all_h3=soup.find_all("h3")

all_countries=[]

for h3 in all_h3:
    name=h3.get_text(strip=True)

    population=h3.find_next("div").select("span.country-population")[0].get_text()

    all_countries.append([name,population])
```

```python
all_countries
```

```python
import pandas as pd

df=pd.DataFrame(all_countries,columns=["Name","Countrues"])

df
```

```python
df.to_csv("cleaned_data/data.csv")
```
